```python
import pandas as pd
dados = pd.read_csv(r'dataset\athlete_events.csv')
dados.head()
```

```python
import matplotlib.pyplot as plt
```

```python
dados.hist(column='Age', bins=100)
plt.show()
```

```python
# altura atletas femininos
atletasFem = dados.loc[dados['Sex'] == 'F']
atletasFem.hist(column='Height', bins=100)
plt.show()
```

```python
# altura atletas masculinos 
atletasMasc = dados.loc[dados['Sex'] == 'M']
atletasMasc.hist(column='Height', bins=100)
plt.show()
```
